```python
import pandas as pd
import nltk
import itertools
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
```

```python
%store -r all_tweets
```

```python
#create the corpus of the tweets
tokens = all_tweets['token']
merge = itertools.chain.from_iterable(tokens)
corpus = list(merge)
```

```python
#import stopwords
stops = set(stopwords.words("english"))
```

```python
#create and visualize the word cloud for top 30 words
wordcloud = WordCloud(stopwords = stops, max_words = 30, scale = 20, background_color = 'White').generate(' '.join(corpus))

plt.figure(figsize=(10,10))
plt.axis("off")
#plt.savefig('plots/word cloud.jpg')
plt.imshow(wordcloud)
myfig = plt.gcf()
myfig.savefig('plots/word cloud.jpg')



#plt.show()
```
